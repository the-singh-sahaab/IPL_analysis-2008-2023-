```python
import numpy as np
import pandas as pd
```

```python
match = pd.read_csv(r"IPL_Matches_2008_2022.csv")
ball = pd.read_csv(r"IPL_Ball_by_Ball_2008_2022.csv")
```

```python
 # we will make the goupyby the matchid and inning to get the total_run in each match by both the team
total_df = ball.groupby(["ID","innings"]).sum()["total_run"].reset_index()
total_df
```

```python
total_df = total_df[total_df["innings"]==1]
```

```python
total_df = total_df[['ID','innings','total_run']]
```

```python
match_df = match.merge(total_df[['ID','total_run']], left_on = "ID",right_on = "ID")
# DOING THIS WE WILL GIVE THE MATCH WALA DATA SET THE COLUMN OF 'TOTAL_RUN' WHIHC WILL TELL THE TOTAL RUN MADE IN THAT ONE INNINGS
```

```python
match_df['Team1'].unique()
```

```python
teams = ['Sunrisers Hyderabad','Mumbai Indians', 'Royal Challengers Bangalore', 'Kolkata Knight Riders', 'Kings XI Punjab', 'Rajasthan Royals', 'Delhi Capitals','Chennai Super Kings']
```

```python
# changing the 'Delhi Daredevils' into 'Delhi Capitals'
match_df['Team1'] = match_df['Team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['Team2'] = match_df['Team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')
# similarly
match_df['Team1'] = match_df['Team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['Team2'] = match_df['Team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df["Team1"].isin(teams)]
match_df = match_df[match_df["Team2"].isin(teams)]

```

```python
match_df.shape
```

NOW WE WILL BRING ONLY THAT COLUMNS WHIHC WILL HELP IN OUR PREDICTIONS


```python
match_df = match_df[["ID","City","WinningTeam",'total_run']]
match_df
filtered_df = match_df[match_df["WinningTeam"] == 'Chennai Super Kings']
filtered_df 
```

```python
ball_df = match_df.merge(ball, on='ID')
ball_df
```

```python
ball_df = ball_df[ball_df["innings"] == 2]
ball_df
```

```python
# now after this we have all the needed thing accept the runs needed and the ball needed and the run required to win the match:
# now to fine this we have to keep the record of run made after each ball and we can o that by using the ball_df data fram's total_run_y
```

```python
# ball_df['currentrun'] = ball_df.groupby("ID").cumsum()['total_run_y']
ball_df['currentrun'] = ball_df.groupby("ID")['total_run_y'].cumsum().astype(int)
ball_df
```

```python
ball_df["runs_left"] = ball_df["total_run_x"] - ball_df["currentrun"]
```

```python
ball_df
```

```python
ball_df["balls_left"] = 126 - (ball_df['overs']*6 + ball_df["ballnumber"])
```

```python
wicket = ball_df.groupby("ID")["isWicketDelivery"].cumsum().astype(int)
ball_df["wicket_left"] = 10 - wicket
ball_df["wicket_left"]
```

```python
# now we will try to calculate the run rate
# run_rate = runs/ overs
ball_df["current_run_rate"] = (ball_df["currentrun"]*6)/(126 - ball_df["balls_left"])
ball_df["required_run_rate"] = (ball_df["currentrun"]*6)/(ball_df["balls_left"])
ball_df
```

```python
# now will make the result of the match column where if the batting tram is winner then 1 else other team wins the its 0
def result(row):
  return 1 if row['BattingTeam'] == row["WinningTeam"] else 0

ball_df["result"] = ball_df.apply(result, axis = 1)
```

```python
ball_df.columns
```

```python
def bal(row):
  return row["Team2"] if row["Team1"] == row["BattingTeam"] else row["Team1"]
```

```python
# Check if 'ID' column exists in both DataFrames
if "ID" in ball_df.columns and "ID" in match.columns:
    # Merge DataFrames on 'ID'
    ball_df = ball_df.merge(match[["ID", "Team1", "Team2"]], on="ID")

    # Check if 'Team1' column exists in the merged DataFrame
    if "Team1" in ball_df.columns:
        ball_df["ballingteam"] = ball_df.apply(bal, axis=1)
    else:
        print("Column 'Team1' not found in the merged DataFrame.")
else:
    print("Column 'ID' not found in one or both DataFrames.")
```

```python
Final_df= ball_df[["ballingteam", "BattingTeam", "City", "balls_left","wicket_left","runs_left","total_run_x","current_run_rate","required_run_rate","result"]]
```

```python
Final_df.isnull().sum()
Final_df.dropna(inplace = True)
Final_df.isnull().sum()
```

```python
# spliting the data
from sklearn.model_selection import train_test_split
y = Final_df.iloc[:,-1]
x = Final_df.iloc[:,:-1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer(
    transformers=[
        ('trf', OneHotEncoder( drop="first"), ["BattingTeam", "ballingteam", "City"])
    ],
    remainder='passthrough'
)
```

```python
# now we will apply logistic regrestion on this data set
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline

pipe = Pipeline(steps = [("step1", trf),("step2",RandomForestClassifier())])
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
